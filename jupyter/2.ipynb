```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data3 = np.random.randn(30).cumsum()
plt.plot(data3, color = 'k', linestyle = 'dotted', label='Default')
plt.plot(data3, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
plt.set_xlabel = "a"
#plt.ylabel = "text b"
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
set = np.random.randn(20).cumsum()
ax.plot(set,color = '#3399ff', label='Default')
ax.plot(set, drawstyle = "steps-post", color = '#ff4d4d', linestyle = '--')
ticks = ax.set_xticklabels([250,500,750,1000,1250,1500])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five', 'six','seven'])
ax.set_title('Noty first matplotlib plot')
ax.set_facecolor('#ffeecc')
ax.legend(loc='best')
ax.set_ylabel('grim')
ax.grid()
ax.set_xlabel('ykr')
```
